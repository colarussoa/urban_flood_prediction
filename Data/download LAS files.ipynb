```python
import os
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin
import os

# Define the HTTPS directory URL
base_url = "https://rockyweb.usgs.gov/vdelivery/Datasets/Staged/Elevation/LPC/Projects/FL_Peninsular_2018_D18/FL_Peninsular_Pinellas_2018/LAZ/"  # Change to your target URL
download_folder = "las/"  # Local folder to save files

# Create download directory if it doesn't exist
os.makedirs(download_folder, exist_ok=True)

def get_file_links(base_url):
    """Fetch all file links from the directory listing page."""
    response = requests.get(base_url)
    
    if response.status_code != 200:
        print("Failed to access URL:", base_url)
        return []

    soup = BeautifulSoup(response.text, "html.parser")
    file_links = []

    # Look for <a> tags with href attributes
    for link in soup.find_all("a", href=True):
        file_url = urljoin(base_url, link["href"])  # Create full URL
        if file_url.endswith(("/")):  # Skip subdirectories (optional)
            continue
        file_links.append(file_url)

    return file_links

def download_file(url, folder):
    """Download a single file."""
    filename = os.path.join(folder, os.path.basename(url))
    response = requests.get(url, stream=True)

    if response.status_code == 200:
        with open(filename, "wb") as file:
            for chunk in response.iter_content(chunk_size=1024):
                file.write(chunk)
        print(f"Downloaded: {filename}")
    else:
        print(f"Failed to download {url}")

# Get all files from the directory listing
file_urls = get_file_links(base_url)

#list of file_url to omit based on strings
omit_str_list = [
    "?",
    "privacy",
    "policies",
    "accessibility-and-us-geological-survey",
    "sitemap",
    "the-national-map-data-delivery/gis-data-download",
    "lesson-4b-downloading-maps-and-data-national-map-download-application",
    "no-fear-act",
    "science-support/foia"
]

#walk through previously downloaded files
folder_path = "./las"  
already_downloaded_files = []
for root, dirs, files in os.walk(folder_path):
    for file in files:
        already_downloaded_files.append(file)
        # print(os.path.join(root, file))

# Download all files
for file_url in file_urls:
    if not any(item in file_url for item in omit_str_list):
        file_name_alone = file_url.split('/')
        if not any(item in file_name_alone[-1] for item in already_downloaded_files):
            download_file(file_url, download_folder)

print("All files downloaded successfully!")
```
